```python
cols_outliers = [
    "app_usage_time_min",
    "likes_received",
    "message_sent_count",
    "mutual_matches"
]
```

```python
import matplotlib.pyplot as plt

for col in cols_outliers:
    plt.figure()
    plt.boxplot(df_main[col], vert=False)
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

```python
def detect_iqr_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[col] < lower) | (data[col] > upper)]
```

```python
for col in cols_outliers:
    outliers = detect_iqr_outliers(df_main, col)
    print(col, "outliers:", len(outliers))
```

```python
common_cols = set(df_main.columns) & set(df_ext.columns)
only_in_main = set(df_main.columns) - set(df_ext.columns)
only_in_ext = set(df_ext.columns) - set(df_main.columns)

print("Common columns:", len(common_cols))
print("Only in df_main:", only_in_main)
print("Only in df_ext:", only_in_ext)
```

```python
mismatch_cols = []

for col in common_cols:
    equal = df_main[col].equals(df_ext[col])
    if not equal:
        mismatch_cols.append(col)

mismatch_cols
```

```python
df_ext['is_mutual_match'] = (df_ext['match_outcome'] == 'Mutual Match').astype(int)


behavior_cols = [
    'app_usage_time_min',
    'swipe_right_ratio',
    'likes_received',
    'message_sent_count',
    'profile_pics_count',
    'bio_length',
    'emoji_usage_rate',
	'last_active_hour'
]


means = df_ext.groupby('is_mutual_match')[behavior_cols].mean().T
means.columns = ['Not Mutual Match', 'Mutual Match']

means['diff'] = means['Mutual Match'] - means['Not Mutual Match']
means = means.sort_values('diff')

plt.figure(figsize=(10, 6))
plt.barh(means.index, means['diff'], color='steelblue')
plt.axvline(0, color='black', linewidth=0.8)

plt.title('Behavioral Feature Differences (Mutual Match − Others)')
plt.xlabel('Mean Difference')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
df_ext['is_ghosted'] = (df_ext['match_outcome'] == 'Ghosted').astype(int)


df_ext['msg_bin'] = pd.cut(
    df_ext['message_sent_count'],
    bins=[0, 5, 20, 50, 100],
    include_lowest=True
)

ghost_rate = df_ext.groupby('msg_bin')['is_ghosted'].mean()

ghost_rate.plot(kind='bar', figsize=(8, 5))
plt.title('Ghosting Rate by Message Count Group')
plt.ylabel('Ghosting rate')
plt.xlabel('Message count range')
plt.tight_layout()
plt.show()
```

```python

df_interest = df_ext[['interest_tags', 'match_outcome']].dropna()
df_interest = df_interest.assign(
    interest=df_interest['interest_tags'].str.split(', ')
).explode('interest')

df_interest['is_mutual_match'] = (
    df_interest['match_outcome'] == 'Mutual Match'
).astype(int)

interest_counts = df_interest['interest'].value_counts()
valid_interests = interest_counts[interest_counts >= 100].index

interest_match_rate_filtered = (
    df_interest[df_interest['interest'].isin(valid_interests)]
    .groupby('interest')['is_mutual_match']
    .mean()
    .sort_values(ascending=False)
)

top_interests = interest_match_rate_filtered.head(10)

plt.figure(figsize=(8, 5))
plt.barh(top_interests.index, top_interests.values)
plt.xlabel("Mutual Match Rate")
plt.title("Top Interests by Mutual Match Rate (min 100 users)")
plt.gca().invert_yaxis()
plt.show()
```
